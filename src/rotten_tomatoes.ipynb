```python
import os

dataset_path = "/home/vscode/.cache/kagglehub/datasets/stefanoleone992/rotten-tomatoes-movies-and-critic-reviews-dataset/versions/1"
print(os.listdir(dataset_path))
```

```python
import pandas as pd

rotten_tomatoes_critic_reviews = pd.read_csv(os.path.join(dataset_path, "rotten_tomatoes_critic_reviews.csv"))
rotten_tomatoes_movies = pd.read_csv(os.path.join(dataset_path, "rotten_tomatoes_movies.csv"))
```

```python
save_path = "../data/raw"
os.makedirs(save_path, exist_ok=True)

rotten_tomatoes_critic_reviews.to_csv(
    os.path.join(save_path, "rotten_tomatoes_critic_reviews.csv"),
    index=False
)

rotten_tomatoes_movies.to_csv(
    os.path.join(save_path, "rotten_tomatoes_movies.csv"),
    index=False
)

print("Files saved in:", save_path)
```

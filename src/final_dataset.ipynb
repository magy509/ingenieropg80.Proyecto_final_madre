```python
import pandas as pd
import random

# Tus datasets
# amazon_df = pd.read_csv("amazon.csv")
# movies_df = pd.read_csv("movies.csv")

# Obtener listas Ãºnicas
unique_product_ids = amazon_df['productId'].unique()
movie_ids = movielens_df['movieId'].tolist()

# Barajamos los movieIds para asignar aleatoriamente
random.shuffle(movie_ids)

# Creamos un diccionario de mapeo productId -> movieId
mapping = {pid: movie_ids[i % len(movie_ids)] for i, pid in enumerate(unique_product_ids)}

# Agregar la columna con movieId al dataset de Amazon
amazon_df['movieId'] = amazon_df['productId'].map(mapping)

# (opcional) guardarlo
# amazon_df.to_csv("amazon_mapped.csv", index=False)

print("Ejemplo de mapeo creado:")
print(amazon_df[['productId', 'movieId']].head(15))
```
